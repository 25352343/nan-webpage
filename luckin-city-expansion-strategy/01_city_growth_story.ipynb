## Purpose of This Analysis

The first step of the expansion analysis is to identify which cities in East China show **real growth potential**.  
Instead of looking only at absolute GMV, we focus on **YoY growth** in the same quarter to reduce seasonality effects.  
We also include **absolute GMV increase** and **store count change** to avoid small-base bias and to support actionable prioritization.



## Business Question & Metrics

### Core Question
Which cities achieved the fastest GMV growth in **22Q1 (Jan–Mar 2022)** compared with **21Q1 (Jan–Mar 2021)**?

### Key Metrics
- **GMV (Gross Merchandise Value)**
- **YoY GMV Growth (22Q1 vs 21Q1)**
- **ΔGMV (absolute GMV increase)**
- **Active store count per city** (and YoY change)

### Why Q1 vs Q1?
Comparing the same quarter helps **reduce seasonality effects** (e.g., holidays and weather) and makes growth more comparable across years.



## Analytical Approach

We aggregate city-level GMV and active stores for:
- **21Q1 (Jan–Mar 2021)**
- **22Q1 (Jan–Mar 2022)**

Key logic:
- Compute **YoY GMV growth** and **ΔGMV**
- Track **store count changes** to support growth diagnosis
- Apply a base GMV threshold in 21Q1 to avoid small-base outliers

SQL used:
- `sql/01_city_growth_q1_yoy.sql`

All queries were executed in **Alibaba Cloud MaxCompute**, and results were reviewed online.



## Key Findings (Summary)

- Some cities show strong YoY growth and also meaningful **ΔGMV**, indicating scalable demand signals.  
- High-GMV cities do not always have high YoY growth, so expansion priority should not rely on size alone.  
- Store count changes vary by city, suggesting growth drivers are different (expansion-driven vs productivity-driven).



## Strategic Implications

Cities with strong YoY growth and meaningful scale (ΔGMV) are shortlisted for deeper analysis.  
Next, we will check whether growth is sustainable at the store level (e.g., **GMV per store**, contribution of new stores vs existing stores).
